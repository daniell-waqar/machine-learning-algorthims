```python
import pandas as pd
from sklearn.datasets import load_breast_cancer  #dataset from sklea
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = load_breast_cancer()
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

```python
X = dataset.copy()   # all columns except target
y = data['target']   # target attribute

# spliting data into training and testing 
X_train,X_test , y_train, y_test = train_test_split(X,y, test_size=0.33)
```

```python
clf = DecisionTreeClassifier(ccp_alpha=0.01)  #tree is growing growing so max_depth is stoping condition
clf.fit(X_train,y_train)    # ccp_alpha is prune that provide good visualization of tree
```

```python
clf.get_params()
```

```python
X_test
```

```python
predictions = clf.predict(X_test)
predictions        # 1 means it's benign and 0 means it's malignant
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)    
# when class is zero and it's predicted as Zero it happens 64 times
# when class is one and it's predicted as one it happens 108 times
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
clf = DecisionTreeClassifier(criterion="gini", max_depth=3)

clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
print("Accuracy : " ,accuracy_score(y_test, y_pred))
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)
print("Accuracy : " ,accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['malignant', 'benign']))
```
